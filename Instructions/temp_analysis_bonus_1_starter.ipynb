### Compare June and December data across all years 

```python
from scipy import stats
```

```python
grouped_stations = df.groupby('station').mean()[["prcp","tobs"]]
grouped_stations
```

```python
# Filter data for desired months
jun_df = df.loc[df['month'] == 6]
jun_df
dec_df = df.loc[df['month'] == 12]
```

```python
# Identify the average temperature for June
grouped_jun_df = jun_df.groupby('station').mean()["tobs"]
grouped_jun_df
```

```python
# Identify the average temperature for December
avg_dec_temp = round((dec_df.mean()["tobs"]),2)
avg_dec_temp
grouped_dec_df = dec_df.groupby('station').mean()["tobs"]
grouped_dec_df
Dec_DF = pd.DataFrame(grouped_dec_df)
Dec_DF
```

```python
# Create collections of temperature data
merged_df = pd.merge(grouped_dec_df, grouped_jun_df, on = 'station', how = "left")

merged_df = merged_df.rename(columns = {'tobs_x':"Dec. Avg. Temp", "tobs_y":"June Avg. Temp"})
merged_df
```

```python
# Run paired t-test
# stats.ttest_rel(df['bp_before'], df['bp_after'])
st.ttest_rel(merged_df["Dec. Avg. Temp"], merged_df["June Avg. Temp"])


```

### Analysis

Hypothesis: There is no significant difference in temperature average in Hawaii in June versus December.  (H0 = 0).

My analysis using a PAIRED T-test indicates there is a statistically significant difference of Hawaii temperatures between June and December, with December being ~7 degrees colder than in June.  A paired t-test was used because the same entities (stations) were measured twice (averages from June and December.)  Because the p-value is less than 0.05, we can conclude that the results are statistically significant, so we reject the null hypothesis and conclude that there IS a significant difference in temperature between June and December in Hawaii based on the results of our weather stations.
